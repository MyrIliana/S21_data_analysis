```python
import pandas as pd
import numpy as np
```

```python
fn= "hypoxiaDataset.complete.csv"
```

```python
data= pd.read_csv(fn, sep=",", header=0)
```

```python
data.head()
```

```python
data[["Id", "Gene_Symbol"]] = data["Id"].str.split("_", 1, expand=True)
```

```python
data = data.set_index("Gene_Symbol")
```

```python
data.head()
```

```python
data["FC"] = data["0.1-O2"] / data["5-O2"]
```

```python
data["log2FC"]= np.log2(data["FC"])
```

```python
data.head()
```

```python
data2 = data.loc[(abs(data["log2FC"]) >= 1) & (abs(data["padj"]) < 0.05)]
```

```python
data2.head()
```

```python
data3 = data2.loc[:,["norm.5.0-O2-n1", "norm.5.0-O2-n2","norm.0.1-O2-n1","norm.0.1-O2-n2", "norm.0.5-O2-n1",
                    "norm.0.5-O2-n2", "norm.1.0-O2-n1","norm.1.0-O2-n2" ]]
```
